```python
import random
import numpy as np
import matplotlib.pyplot as plt


allnum=[]
percent=[]

for l in range(10,90):
    thisnum=[]
    streak=1
    percent.append(l)
    for i in range(1000000):
        random_number = random.random()
        if random_number<(l/100):
            streak = streak +1
        else:
            streak = 1
        thisnum.append(streak)
    allnum.append(thisnum)
data=np.mean(allnum,axis=1)  
```

```python
plt.plot(percent,data)
plt.ylabel('Mean hours on same source cell')
plt.xlabel('Percent chance for selection')
plt.xticks(np.arange(min(percent), max(percent)+1, 10.0))
plt.show()
```

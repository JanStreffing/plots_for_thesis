```python
from PIL import Image, ImageDraw, ImageFont
import matplotlib.pyplot as plt

# Set the dimensions of the image
width = 1800  # Total width
height = 80  # Height of the image

# Number of dots and their colors (customize this list)
total_dots = 34
section_borders = [0,4,11,24,30,33]
section_titles = ['Foundations', 'Base Model', 'Deterministic', 'Stochastic', 'Outlook','']

font = ImageFont.truetype("arial.ttf", size=18)  # You can change the font and size


# Create a new blank image
image = Image.new("RGB", (width, height), color="white")
draw = ImageDraw.Draw(image)

dot_position =[]

for dot_index in range(total_dots):
    dot_position.append((width-width/total_dots)/total_dots*dot_index+width/total_dots)

already_drawn=0
# Loop to create the images
for page_index in range(total_dots):
    
    # Clear the previous content by filling with white
    draw.rectangle([0, 0, width, height], fill="white")
    
    for dot_index in range(total_dots):

        # Draw dots in all sections
        if already_drawn == dot_index:
            dot_color = "navy"
        elif already_drawn > dot_index:
            dot_color = "slategrey"
        else:
            dot_color = "lavender"

        # Get the position of the dot within the section
        dot_x = dot_position[dot_index]
        dot_y = height // 1.6
        
        if dot_index in section_borders:
            # Calculate the top and bottom positions of the line
            line_top = dot_y + 30  # 15 pixels above the center
            line_bottom = dot_y - 30  # 15 pixels below the center

            # Draw a vertical line centered on the image
            draw.line((dot_x, line_top, dot_x, line_bottom), fill="slategrey", width=4)
            
            # Add section titles as text
            section_title = section_titles[section_borders.index(dot_index)]
            
            text_width, text_height = draw.textsize(section_title, font=font)
            text_x = dot_x+10 
            text_y = line_bottom - text_height - 5  # Adjust position
            draw.text((text_x, text_y), section_title, fill="black", font=font)

        else:
            draw.ellipse(
                [(dot_x - 10, dot_y - 10),
                 (dot_x + 10, dot_y + 10)],
                fill=dot_color
        )
    already_drawn=already_drawn+1

    # Crop the image to remove white borders
    bbox = image.getbbox()
    cropped_image = image.crop(bbox)

    # Save the cropped image
    cropped_image.save(f"progressbar/image_{page_index + 1}.png")


print("Images generated successfully.")
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import (MultipleLocator, FormatStrFormatter, AutoMinorLocator)

# Define the time scales for each component
fesom1_scales = [(10, 200)]
fesom2_scales = [(1, 120)]
echam6_scales = [(100, 200)]  # Corrected to have a width of 100
openifs43_scales = [(10, 120)]

# Create the plot
fig, ax = plt.subplots(figsize=(6, 2))

height = 0.7

xmin = 0
xmax = 250
x_values = range(xmin, xmax)

# Plot each component as a horizontal bar
ax.barh(1, fesom1_scales[0][1] - fesom1_scales[0][0], left=fesom1_scales[0][0], height=height, align='center', color='red')
ax.barh(2, fesom2_scales[0][1] - fesom2_scales[0][0], left=fesom2_scales[0][0], height=height, align='center', color='orange')
ax.barh(3, echam6_scales[0][1] - echam6_scales[0][0], left=echam6_scales[0][0], height=height, align='center', color='blue')
ax.barh(4, openifs43_scales[0][1] - openifs43_scales[0][0], left=openifs43_scales[0][0], height=height, align='center', color='darkcyan')

# Set the x and y labels
ax.set_xlabel('Characteristic spatial scales in km')
ax.set_yticks([1, 2, 3, 4])
ax.set_xlim(0, 250)  # Set the x-axis limits
ax.set_yticklabels(['FESOM1', 'FESOM2', 'ECHAM6', 'OpenIFS43'])

# Save the plot as an image
plt.savefig('resrange.png', dpi=300, bbox_inches='tight')
```

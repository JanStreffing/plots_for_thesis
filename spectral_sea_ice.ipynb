```python
import matplotlib.pyplot as plt
import numpy as np

wavelength = ['185','250', '440', '690', '1190', '2380', '4000']
data = [1.12114,1.12114, 1.12463, 1.12561, 1.08982, 0.67650, 0.14922]
weight = [0.001917,0.001917, 0.135708, 0.322135, 0.326158, 0.180608, 0.033474]
labels = wavelength

data_array = np.array(data)

data1 = data_array * 0.8
data2 = data_array * 0.7
data3 = data_array * 0.6
data4 = data_array * 0.5

# Set figure size
plt.figure(figsize=(6, 4))

plt.fill_between(labels, weight, step='post', color='grey', alpha=.5, label='band weight',zorder=0)
plt.step(labels, data1, color='b', label='α=0.8')
plt.step(labels, data2, color='r', label='α=0.7')
plt.step(labels, data3, color='g', label='α=0.6')
plt.step(labels, data4, color='orange', label='α=0.5')
plt.xlabel('Wavelength (nm)', fontsize=12)
plt.ylabel('Spectral albedo', fontsize=12)
plt.title('Broadband to spectral albedo mapping')
plt.legend()


# Set y-axis label font size
#plt.yticks(fontsize=10)

# Save plot
plt.savefig('spectral_mapping.png',dpi=300)

plt.show()
```
